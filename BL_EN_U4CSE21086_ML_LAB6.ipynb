```python
#Defining a text PreProcessing Functions

import string
import re
def text_processing(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
data['news']=data['news'].fillna('').apply(text_processing)
```

```python
x=data['news']
y=data['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector=TfidfVectorizer()
X_train=vector.fit_transform(x_train)
X_test=vector.transform(x_test)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
LR = LogisticRegression()
LR.fit(X_train,y_train)
pred_lr=LR.predict(X_test)
map_label={0:'fake',1:'not fake'}
pred=[map_label[p] for p in pred_lr]
data_test=x_test.copy().reset_index()
data_test['bin_pred']=pred
data_test.drop('index',axis=1,inplace=True)
LR.score(X_test, y_test)
print("predictions are ")
print(data_test[['news','bin_pred']])
```

```python
accuracy=accuracy_score(y_test,pred_lr)
print("accuracy is ",accuracy)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.feature_extraction.text import TfidfVectorizer

# Load your dataset from 'train_news.csv'
data = pd.read_csv('train_news.csv')

# Drop rows with missing values in the 'news' and 'label' columns
data = data.dropna(subset=['news', 'label'])

x = data['news']
y = data['label']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)

# TF-IDF feature extraction
vectorizer = TfidfVectorizer(max_features=2000)
x_train_tfidf = vectorizer.fit_transform(x_train)
x_test_tfidf = vectorizer.transform(x_test)

reg_tree = DecisionTreeRegressor(random_state=50)
reg_tree.fit(x_train_tfidf, y_train)
y_pred_regtree = reg_tree.predict(x_test_tfidf)

error_regtree = mean_squared_error(y_test, y_pred_regtree)
print("Mean squared error of regression tree is", error_regtree)

threshold = 0.5

# Converting the predictions generated by the model into fake or not fake binary class predictions
bin_pred = ["not fake" if p >= threshold else "fake" for p in y_pred_regtree]
data_test = data.loc[y_test.index]
data_test['bin_pred_regtree'] = bin_pred
error_binary = mean_squared_error(y_test, [1 if p == "not fake" else 0 for p in bin_pred])
print("Mean squared error for binary predictions is", error_binary)
print("Predictions are:")
print(data_test[['news', 'bin_pred_regtree']])

```
